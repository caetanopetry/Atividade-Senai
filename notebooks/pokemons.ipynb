## 4Ô∏è‚É£ Estat√≠sticas simples
- Calcular a m√©dia da coluna Total
- Descobrir o maior Total
- Descobrir o menor Total
- Calcular a m√©dia de Attack dos Pok√©mons da Gera√ß√£o 1
- Calcular a m√©dia de Attack dos Pok√©mons da Gera√ß√£o 6
- Calcular a m√©die de HP dos Pok√©mons do tipo Fire
- Calcular a m√©die de HP dos Pok√©mons do tipo Water

```python

media_colum_total = df["Total"].mean()
print(f"M√©dia da coluna total: {media_colum_total}")

maior_total = df["Total"].max()
print(f"Maior Total: {maior_total}")

menor_total = df["Total"].min()
print(f"Maior Total: {menor_total}")

media_attack_gen_1= df[df["Generation"]==1]["Attack"].mean()
print(f"Media de Attack dos Pokem√≥ns da gera√ß√£o 1: {media_attack_gen_1:.2f} ")

media_attack_gen_6= df[df["Generation"]==6]["Attack"].mean()
print(f"Media de Attack dos Pokem√≥ns da gera√ß√£o 6: {media_attack_gen_6:.2f} ")

media_hp_fire = df[df["Type 1"]== "Fire"]["HP"].mean()
print(f"M√©dia de HP dos Pok√©mons do tipo Fire: {media_hp_fire:.2f}")

media_hp_water= df[df["Type 1"]== "Water"]["HP"].mean()
print(f"M√©dia de HP dos Pok√©mons do tipo Water: {media_hp_water:.2f}")
```

## 5Ô∏è‚É£ Ordena√ß√£o
- Ordenar os Pok√©mon por Total do maior para o menor.
- Ordenar os Pok√©mon por Speed do menor para o maior.

```python
ordernartotal = df.sort_values(by="Total", ascending=True)
ordernartotal

ordernarspeed = df.sort_values(by="Speed", ascending=True)
ordernarspeed
```
